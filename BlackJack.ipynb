```python
from random import shuffle
from IPython.display import clear_output
gameon = False

#defining the attributes needed in some classes globally
suits = ['Hearts', 'Diamonds', 'Spades', 'Clubs']
ranks = ['Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace']
values = {'Two':2 , 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}
```

```python
#returns a card in a deck
class Cards:
    
    def __init__(self, suit, rank):
        
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
    
    def __str__(self):
        return (f'{self.rank} of {self.suit}')
```

```python
#Creating a 52 card Deck with a shoe value. Shoe is the amount of Decks to be created
class Deck:
    
    def __init__(self,shoes):
        self.allcards = []
        self.shoes = shoes
        count=0
        while count<self.shoes:
            for suit in suits:
                for rank in ranks:
                    self.allcards.append(Cards(suit,rank))
            count += 1
    
    def __str__(self):
        return(f"The table is playing with {self.shoes} shoes containing {self.shoes*52} cards")
    
    def shuffle(self):
        shuffle(self.allcards)
    
    #To deal out cards to both dealer and player
    def dealcards(self):
        return self.allcards.pop()
```

```python
class Player:
    
    def __init__(self,name,balance):
        self.name = name
        self.balance = balance
        self.playercards = []
        
    def __str__(self):
        return (f"{self.name} has {self.balance} dollars")
    
    def win(self,wins):
        self.balance += wins
        
    def lose(self,loss):
        self.balance -= loss
        
    #To topup player amount if player becomes low on cash
    def topup(self):
        topup = 'WRONG'
        while topup not in ['Y','N']:
            topup = input('Do you want topup? (Y or N): ')
            topup = topup.upper()
        if topup == 'Y':
            amount = 'WRONG'
            while amount.isdigit() == False:
                amount = input('Amount(10-1000): ')
                self.balance +=int(amount)
        else:
            return False
        
    #To display the cards a player has, at an given point.
    def display(self):
        print(f'\n{self.name} Cards are : ')
        for card in self.playercards:
            print(card)
```

```python
#Counts the values of cards that has a player/dealer has.
def count(playinglist):
    countval = 0
    for val in playinglist:
        countval += val.value
    return countval
```
